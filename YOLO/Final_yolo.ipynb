* Save the result as a video
* print number of objects
* print speed(Detection Time) Cpu or GPU







```python
import torch
import cv2
```

```python
%pip install ultralytics
from ultralytics import YOLO
```

```python
input_path = '/content/Test_sample_video.mp4'
output_path = '/content/output/Test_sample_video.mp4'
```

```python
def YoloModel(model,path,path2):
  video = cv2.VideoCapture(path)
  width = int(video.get(3))
  height = int(video.get(4))
  #video encoding:
  fourcc = cv2.VideoWriter_fourcc(*'mp4v')
  encoder = cv2.VideoWriter(path2, fourcc, 25, (width, height))
  Speed = 0
  Number_of_objects = 0
  while 1:
     ret, frame = video.read()
     if not ret :
       break
     results = model.predict(frame)[0]
     image= cv2.cvtColor(results.plot(), cv2.COLOR_BGR2RGB)
     encoder.write(image)
     Speed +=  sum(results.speed.values())
     Number_of_objects += len(results.boxes.data)
  video.release()
  encoder.release()
  print(f'Number of objects Detected : {Number_of_objects}, speed(Detection Time) {Speed}')
```

#Yolov3

```python
model = YOLO('yolov3u.pt')
YoloModel(model,input_path,output_path)
```

#Yolov5

```python
model = YOLO('yolov5mu.pt')
YoloModel(model,input_path,output_path)
```
